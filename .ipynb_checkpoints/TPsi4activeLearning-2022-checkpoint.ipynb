### let's compare SVM, with different classification models

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.cluster import KMeans

def evaluate_classifier(classifier, x_train, y_train, x_test, y_test, display_labels):
    start_time = time.time()

    # Fit the model
    classifier.fit(x_train, y_train)
    
    # Make predictions
    y_pred = classifier.predict(x_test)
    
    # Calculate metrics
    accuracy = accuracy_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    
    end_time = time.time()
    elapsed_time = end_time - start_time
    
    # Display the metrics
    print(f"Accuracy score: {accuracy}")
    print(f"F1 score: {f1}")
    print(f"Time taken for prediction and scoring: {elapsed_time} seconds")
    
    # Display the confusion matrix
    cm = confusion_matrix(y_test, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=display_labels)
    disp.plot(cmap='Blues')
    plt.show()
```

```python
# SVM Classifier
print("=== SVM Classifier ===")
svm_classifier = svm.SVC(kernel='linear', C=1.0)
evaluate_classifier(svm_classifier, x_train, y_train, x_test, y_test, display_labels=[3, 7])

# Decision Tree Classifier
print("=== Decision Tree Classifier ===")
tree_classifier = DecisionTreeClassifier()
evaluate_classifier(tree_classifier, x_train, y_train, x_test, y_test, display_labels=[3, 7])

# Logistic Regression Classifier
print("=== Logistic Regression Classifier ===")
logistic_classifier = LogisticRegression(max_iter=1000)
evaluate_classifier(logistic_classifier, x_train, y_train, x_test, y_test, display_labels=[3, 7])
```

## OVO and OVR classifcation of 3 classes

### we choose the 3 classes 0, 1 and 2

```python
from tensorflow.keras.datasets import mnist

(x_train, y_train), (x_test, y_test) = mnist.load_data()

indices_train = (y_train == 0) | (y_train == 1) | (y_train == 2)
indices_test = (y_test == 0) | (y_test == 1) | (y_test == 2)

x_train = x_train[indices_train]
y_train = y_train[indices_train]
x_test = x_test[indices_test]
y_test = y_test[indices_test]

x_train = x_train.reshape(-1, 28*28)
x_test = x_test.reshape(-1, 28*28)

scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

### One vs one SVM classification

```python
from sklearn.multiclass import OneVsOneClassifier


start_time = time.time()
ovo_classifier = OneVsOneClassifier(svm.SVC())
ovo_classifier.fit(x_train, y_train)
ovo_y_pred = ovo_classifier.predict(x_test)
ovo_accuracy = accuracy_score(y_test, ovo_y_pred)
ovo_time = time.time() - start_time

print(f"OvO Accuracy: {ovo_accuracy}, Time taken: {ovo_time} seconds")
cm = confusion_matrix(y_test, ovo_y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot(cmap='Blues')
plt.show()
```

### One vs Rest SVM classification

```python
from sklearn.multiclass import OneVsRestClassifier

start_time = time.time()
ovr_classifier = OneVsRestClassifier(svm.SVC())
ovr_classifier.fit(x_train, y_train)
ovr_y_pred = ovr_classifier.predict(x_test)
ovr_accuracy = accuracy_score(y_test, ovr_y_pred)
ovr_time = time.time() - start_time

print(f"OvR Accuracy: {ovr_accuracy}, Time taken: {ovr_time} seconds")
cm = confusion_matrix(y_test, ovr_y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot(cmap='Blues')
plt.show()
```

## Active learning with SVM

```python
from tensorflow.keras.datasets import mnist
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Reshape training data
x_train = x_train.reshape(-1, 28*28)

# Reshape test data
x_test = x_test.reshape(-1, 28*28)

# Get the indices of the elements that are either 3 or 7
indices_train = (y_train == 3) | (y_train == 7)
indices_test = (y_test == 3) | (y_test == 7)

# Filter the data based on the indices
x_train = x_train[indices_train]
x_test = x_test[indices_test]

# Filter the labels and set them to 0 for class '3' and 1 for class '7'
y_train = np.where(y_train[indices_train] == 3, 0, 1)
y_test = np.where(y_test[indices_test] == 3, 0, 1)
```

**Question 6: Initialisation of the active training dataset**

Construct a new training dataset named (xActif,yActif). For it's initialisation, take randomly nb0 data from the copy of the original training dataset (xTrainP, yTrainP). You are allowed to use informations from yTrainP in order to get half of nb0 data for each class. These nb0 data are also removed from (xTrainP,yTrainP). Removing data can be done using [np.delete](https://numpy.org/doc/stable/reference/generated/numpy.delete.html).

```python
# we assume that nb0 is an even number
nb0 = 4 # number of data in the active training dataset at initialisation
xActif = []
yActif = []
```

```python
#your answer
# Get indices for each class
indices_class0 = np.where(yTrainP == 0)[0]
indices_class1 = np.where(yTrainP == 1)[0]

# Randomly sample half of nb0 from each class
random_indices_class0 = np.random.choice(indices_class0, nb0 // 2, replace=False)
random_indices_class1 = np.random.choice(indices_class1, nb0 // 2, replace=False)

# Add the sampled data to xActif and yActif
xActif = np.concatenate([xTrainP[random_indices_class0], xTrainP[random_indices_class1]])
yActif = np.concatenate([yTrainP[random_indices_class0], yTrainP[random_indices_class1]])

# Remove the sampled indices from the original datasets
xTrainP = np.delete(xTrainP, np.concatenate([random_indices_class0, random_indices_class1]), axis=0)
yTrainP = np.delete(yTrainP, np.concatenate([random_indices_class0, random_indices_class1]), axis=0)
```

**Question 7: Iterations of the active learning** 

1. Learn a linear SVM classifier on the active training dataset
2. Compute the accuracy on the test dataset (not modified)
3. add randomly nb new data to the active training dataset and remove them from (xTrainP, yTrainP)
4. Go back to step 1 (20 times)

```python
from sklearn import svm
from sklearn.metrics import accuracy_score

# Let's assume nb is the number of new data points to add in each iteration
nb = 2  # You can set it to any number you'd like
accuracy_list = []
for i in range(20):  # 20 iterations
    # Step 1: Train a linear SVM on the active training dataset
    classifier = svm.SVC(kernel='linear')
    classifier.fit(xActif, yActif)
    
    # Step 2: Compute the accuracy on the test dataset
    y_pred = classifier.predict(x_test)
    accuracy = accuracy_score(y_test, y_pred)
    accuracy_list.append(accuracy)
    
    # Step 3: Add nb new data points to the active dataset and remove them from the original one
    # Randomly sample nb indices from the remaining training set
    random_indices = np.random.choice(len(xTrainP), nb, replace=False)
    
    # Add these to the active training set
    xActif = np.concatenate([xActif, xTrainP[random_indices]])
    yActif = np.concatenate([yActif, yTrainP[random_indices]])
    
    # Remove these from the original training set
    xTrainP = np.delete(xTrainP, random_indices, axis=0)
    yTrainP = np.delete(yTrainP, random_indices, axis=0)
```

Plot the accuracy with respect to the iterations from the previous question.

```python
plt.figure()
plt.plot(range(1, 21), accuracy_list)  # x-values are the iteration numbers (1 to 20)
plt.xlabel('Iteration')
plt.ylabel('Test Accuracy')
plt.title('Evolution of Test Accuracy Over Iterations')
plt.show()
```

**Question 9: strategy for choosing new data**
    
Same question as question 7 but, instead of choosing the new points randomly, at each iteration, choose the nb points that are the closest to the separation. The [decision_function](https://scikit-learn.org/stable/modules/generated/sklearn.svm.SVC.html#sklearn.svm.SVC.decision_function) from scikit-learn will help you.

```python
def active_learning_random(xTrainP, yTrainP, x_test, y_test, nb0, nb, iterations):
    accuracy_list_random = []
    random_indices = np.random.choice(len(xTrainP), nb0, replace=False)
    xActif = xTrainP[random_indices]
    yActif = yTrainP[random_indices]
    xTrainP = np.delete(xTrainP, random_indices, axis=0)
    yTrainP = np.delete(yTrainP, random_indices, axis=0)

    for i in range(iterations):
        classifier = svm.SVC(kernel='linear')
        classifier.fit(xActif, yActif)
        y_pred = classifier.predict(x_test)
        accuracy = accuracy_score(y_test, y_pred)
        accuracy_list_random.append(accuracy)
        random_indices = np.random.choice(len(xTrainP), nb, replace=False)
        xActif = np.vstack([xActif, xTrainP[random_indices]])
        yActif = np.concatenate([yActif, yTrainP[random_indices]])
        xTrainP = np.delete(xTrainP, random_indices, axis=0)
        yTrainP = np.delete(yTrainP, random_indices, axis=0)
    return accuracy_list_random

def active_learning_closest(xTrainP, yTrainP, x_test, y_test, nb0, nb, iterations):
    accuracy_list_closest = []
    random_indices = np.random.choice(len(xTrainP), nb0, replace=False)
    xActif = xTrainP[random_indices]
    yActif = yTrainP[random_indices]
    xTrainP = np.delete(xTrainP, random_indices, axis=0)
    yTrainP = np.delete(yTrainP, random_indices, axis=0)

    for i in range(iterations):
        classifier = svm.SVC(kernel='linear')
        classifier.fit(xActif, yActif)
        y_pred = classifier.predict(x_test)
        accuracy = accuracy_score(y_test, y_pred)
        accuracy_list_closest.append(accuracy)
        decision_values = np.abs(classifier.decision_function(xTrainP))
        closest_indices = np.argsort(decision_values)[:nb]
        xActif = np.vstack([xActif, xTrainP[closest_indices]])
        yActif = np.concatenate([yActif, yTrainP[closest_indices]])
        xTrainP = np.delete(xTrainP, closest_indices, axis=0)
        yTrainP = np.delete(yTrainP, closest_indices, axis=0)
    return accuracy_list_closest
```

```python
# Load the data
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Reshape the data
x_train = x_train.reshape(-1, 28*28)
x_test = x_test.reshape(-1, 28*28)

# Filter the data to only include 3s and 7s
indices_train = (y_train == 3) | (y_train == 7)
indices_test = (y_test == 3) | (y_test == 7)

x_train = x_train[indices_train]
y_train = np.where(y_train[indices_train] == 3, 0, 1)

x_test = x_test[indices_test]
y_test = np.where(y_test[indices_test] == 3, 0, 1)
```

**Question 10: plot the evolution of the accuracy**

Plot the accuracy with respect to the iterations from the previous question.
Compare with question 8. Also compare with the baseline.

```python
# Number of iterations
iterations = 20

xTrainP_random = np.copy(x_train)
yTrainP_random = np.copy(y_train)

xTrainP_closest = np.copy(x_train)
yTrainP_closest = np.copy(y_train)

accuracy_list_random = active_learning_random(xTrainP_random, yTrainP_random, x_test, y_test, nb0=4, nb=2, iterations=iterations)
accuracy_list_closest = active_learning_closest(xTrainP_closest, yTrainP_closest, x_test, y_test, nb0=4, nb=2, iterations=iterations)

plt.figure(figsize=(10, 6))
plt.plot(range(1, iterations + 1), accuracy_list_random, marker='o', linestyle='-', label='Random Selection')
plt.plot(range(1, iterations + 1), accuracy_list_closest, marker='x', linestyle='--', label='Closest to Separation')
plt.xlabel('Iteration', fontsize=12)
plt.ylabel('Accuracy', fontsize=12)
plt.title('Accuracy Comparison in Active Learning', fontsize=14)
plt.yticks(np.arange(0.5, 1.1, 0.02))  # Setting ticks at intervals of 0.02 starting from 0.5
plt.ylim([0.5, 1.0])  # Setting the limit for the y-axis
plt.grid(True)
plt.legend(loc='lower right', fontsize=12)
plt.show()
```

**Question 11: many random starts**
    
Since the initialisation is random, running previous codes can lead to different curves for questions 8 and 10. Write here the code necessary for plotting several (e.g. 10) curves corresponding to questions 8 and 10 and display these new plots. Which one is the best strategy?
    

```python
import matplotlib.pyplot as plt

# Initialize plotting
plt.figure(figsize=(12, 8))

# Number of trials
num_trials = 10

# Run the random selection strategy multiple times and plot the results
for i in range(num_trials):
    xTrainP = np.copy(x_train)
    yTrainP = np.copy(y_train)
    accuracy_list_random = active_learning_random(xTrainP, yTrainP, x_test, y_test, 4, 2, 20)
    plt.plot(range(1, 21), accuracy_list_random, label=f'Random Trial {i+1}', alpha=0.5)

# Run the closest selection strategy multiple times and plot the results
for i in range(num_trials):
    xTrainP = np.copy(x_train)
    yTrainP = np.copy(y_train)
    accuracy_list_closest = active_learning_closest(xTrainP, yTrainP, x_test, y_test, 4, 2, 20)
    plt.plot(range(1, 21), accuracy_list_closest, label=f'Closest Trial {i+1}', linestyle='--', alpha=0.5)

plt.title("Active Learning: Random vs. Closest Selection")
plt.xlabel("Iteration")
plt.ylabel("Accuracy")
plt.xticks(np.arange(1, 21, 1))
plt.yticks(np.arange(0.5, 1.01, 0.02))
plt.ylim([0.5, 1.0])
plt.legend()
plt.grid(True)
plt.show()
```
